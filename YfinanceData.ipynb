```python
from math import *
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import yfinance as yf

figsize = (10, 5)    # figure size - change depending on your screen size
plt.rcParams['figure.figsize'] = figsize
```

```python
ticker = 'AAPL'
stock = yf.Ticker(ticker)
intraday_data = stock.history(period='1d', interval='1m')  # Получить минутные данные за 1 день

prices = intraday_data['Close']

#print(type(prices))
# В prices теперь содержатся минутные данные о цене закрытия акций за текущий день
#print(prices)

#dictionary = prices.to_dict()

#print(dictionary)

prices.plot(title=f'{ticker} stock prices', grid=True, figsize=figsize)
plt.show()
```

```python
# Compute and plot log-returns
returns = np.log(prices) - np.log(prices.shift(1))
returns.plot(title=f'{ticker} log-returns', grid=True, figsize=figsize,
             ylim=(returns.min()*1.05, returns.max()*1.05))
plt.show()
```
